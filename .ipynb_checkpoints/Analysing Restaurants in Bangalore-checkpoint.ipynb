# Analysing restaurants in Bangalore

Importing all the required necessary libraries to be used for the analysis.

```python
import pandas as pd
import requests
import json
import folium
from pandas.io.json import json_normalize
```

Defining the required parameters to make API calls to the Foursquare API.  

```python
CLIENT_ID = 'XXX'
CLIENT_SECRET = 'YYY'
VERSION = '20180605'
LIMIT = 100
radius = 9000
```

The latitude and longitude of Bangalore.

Visualising the city using the Folium library

This will extract the required category from the fetched data.

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

This will be the URL which will be used to make API calls and fetch the required data received in JSON format.

```python
url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
```

The results variable will store all the data that the API call fetches.

```python
results = requests.get(url).json()
```

The foursquare venues are going to be kept in one place in the fvenues dataframe.

The data needs to be cleaned as it contains a lot of unnecessary information appended to it.

```python
fvenues = pd.DataFrame(columns = ['name', 'categories', 'lat', 'lng'])

no_of_venues = len(results['response']['groups'][0]['items'])

print("{} Venues fetched within a radius of {} Km".format(no_of_venues, radius/1000))


venues = results['response']['groups'][0]['items']
venues = pd.json_normalize(venues)


filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
venues = venues.loc[:, filtered_columns]


venues['venue.categories'] = venues.apply(get_category_type, axis = 1)


#Dropping the prefix "venue" from each column
venues.columns = [col.split(".")[-1] for col in venues.columns]


fvenues = pd.concat([fvenues, venues], axis = 0, sort = False)


fvenues = fvenues.reset_index(drop = True)
fvenues.rename(columns={"name": "Name", "categories": "Categories","lat":"Latitude","lng":"Longitude"},inplace=True)
fvenues
```

We will now fetch the restaurant from the Zomato API with the help of the latitudes and longitudes of the venue already contained in the fvenues dataframe so that we can have the price_for_two and ratings dataset for each restaurant and also enabling us to drop the columns which are not restaurants from the fvenues dataframe.

Just a visualization to show the accuracy of the location from both of the datasets combined.

```python
blr_map = folium.Map(location = [lat, lng], zoom_start = 12)

for name, latitude, longitude in zip(fvenues['Name'], fvenues['Latitude'], fvenues['Longitude']):
    label = '{}'.format(name)
    label = folium.Popup(label, parse_html = True)
    folium.CircleMarker(
        [latitude, longitude],
        radius = 5,
        popup = label,
        color = 'green',
        fill = True,
        fill_color = '#3186cc',
        fill_opacity = 0.7,
        parse_html = False).add_to(blr_map)  

blr_map
```

The latitudes and longitudes are rounded off to the 4th decimal digit here.

```python
fvenues['Latitude'] = fvenues['Latitude'].apply(lambda lat: round(float(lat), 4))
fvenues['Longitude'] = fvenues['Longitude'].apply(lambda lng: round(float(lng), 4))
zvenues['latitude'] = zvenues['latitude'].apply(lambda lat: round(float(lat), 4))
zvenues['longitude'] = zvenues['longitude'].apply(lambda lng: round(float(lng), 4))
```

The combined dataset will contain the difference of the corresponding latitude and longitude values from both of the datasets.

```python
dataset = pd.concat([fvenues, zvenues], axis = 1)
dataset['lat_diff'] = dataset['Latitude'] - dataset['latitude']
dataset['lng_diff'] = dataset['Longitude'] - dataset['longitude']
```

```python
pd.set_option('display.max_rows', None)
```

We are going to remove the places which are not geographically accurate.

An accuracy of 0.0005 is considered neglegible and is thus tolerable (actual difference is only about 55 metres) and considered to be virtually the same place and hence retained in the dataset.

```python
selected_venues = dataset[(abs(dataset['lat_diff']) <= 0.0005) & (abs(dataset['lng_diff']) <= 0.0005)].reset_index(drop = True)
selected_venues
```

Now to manually go through the data to remove the columns which are not restaurants or the restaurants which have a different name for the same latitude and longitude values owing to the disparity in the data form both the APIs

```python
selected_venues = selected_venues.drop([14, 10, 17, 27, 31, 40, 47, 59, 62, 67, 72, 74, 75, 76]).reset_index(drop = True)
selected_venues
```

The features dataframe will be used to cluster the restaurants together and we will be making use of the price_for_two and ratings data from the Zomato API

```python
features = selected_venues[['price_for_two','rating']]
features
```

Importing the necessary libraries/modules:

Clustering - Kmeans
Visualisation - pyplot from matplotlib
Normalizing data - StandardScaler

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

We will normalize the data because the values of ratings range only from 0-5 and price_for_two has a much wider scale.

```python
data = StandardScaler().fit_transform(features)
data
```

We will be using Within Cluster Sum of Squares to identify the correct amount of clusters to be made for the data above.

```python
wcss=[]
for i in range(1,11): 
    kmeans = KMeans(n_clusters=i, init ='k-means++', max_iter=300, random_state=0)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
```

The elbow curve method suggests us to use a cluster of 3 labels

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

So we will fit the data creating 3 clusters for the same and the corresponding labels will be stored into the features dataframe

```python
clustering_kmeans = KMeans(n_clusters=3)
model = clustering_kmeans.fit(data)
features['cluster'] = clustering_kmeans.fit_predict(data)
```

We need to convert the data types to float otherwise the scatter plot will raise an exception

```python
features = features.astype({"rating": float})
features.info()
```

The scatter plot helps us visualize the clusters which were formed by the k-means algorithm above and we can classify the labels as well: 

1.Cluster 0 (colour - green) : These are the restaurants who have a relatively low price and an highly rated (average > 4.5)

2.Cluster 1 (colour - red): These are the restaurants which are cheap but are not rated as well as the ones above so these are the ones to avoid (average rating < 4)

3.Cluster 3 (colour - blue): These are the restaurants which are rated pretty high but are also quite expensive (Average price > 2000â‚¹) 


```python
plt.figure(figsize=(10,8))

plot = plt.scatter(features['price_for_two'], features['rating'], c=model.labels_.astype(float))
label = [0,1,2]
plt.legend(handles=plot.legend_elements()[0], labels=label)
plt.title('K-means Clustering with Price for two and rating for the restaurants')
plt.xlabel("Price for two")
plt.ylabel("Restaurant Rating")
plt.show()
```

Storing these clusters into the original dataframe.

```python
selected_venues['Clusters'] = features['cluster']
selected_venues
```
